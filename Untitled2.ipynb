```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import pmdarima as pm
from prophet import Prophet

print("✅ All libraries imported successfully!")
```

```python
import pandas as pd
import numpy as np

# Generate realistic monthly sales data (2.5 years)
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=30, freq='M')
sales = 10000 + np.cumsum(np.random.normal(200, 300, 30)) + \
        500 * np.sin(np.arange(30) * np.pi / 6)  # Add seasonality

df = pd.DataFrame({'Date': dates, 'Sales': sales})
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

print("✅ Sample Dataset Created!")
print(df.head())
print("\nShape:", df.shape)
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Plot historical trend
plt.figure(figsize=(12, 6))
plt.plot(df['Sales'], marker='o')
plt.title('Historical Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# Decompose into Trend + Seasonal + Residual
decomp = seasonal_decompose(df['Sales'], model='additive', period=12)
decomp.plot()
plt.show()
```

```python
# Use last 6 months as test data
train = df.iloc[:-6]
test = df.iloc[-6:]

print("Training data:", train.shape)
print("Test data:", test.shape)
```

```python
from pmdarima import auto_arima
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ---------------- Auto ARIMA Model ----------------
print("Training Auto ARIMA model...")
model_arima = auto_arima(train['Sales'], seasonal=True, m=12, D=0, trace=False)
forecast_arima = model_arima.predict(n_periods=6)

print("\n✅ ARIMA Forecast done!")

# ---------------- Prophet Model ----------------
print("Training Prophet model...")
df_prophet = df.reset_index().rename(columns={'Date': 'ds', 'Sales': 'y'})
m = Prophet(yearly_seasonality=True)
m.fit(df_prophet.iloc[:-6])
future = m.make_future_dataframe(periods=6, freq='M')
forecast_prophet = m.predict(future)
forecast_prophet_tail = forecast_prophet['yhat'].tail(6).values

print("✅ Prophet Forecast done!")
```

```python
# ---------------- Prophet Model (Cleaned) ----------------
print("Training Prophet model...")
df_prophet = df.reset_index().rename(columns={'Date': 'ds', 'Sales': 'y'})

m = Prophet(yearly_seasonality=True,
            weekly_seasonality=False,
            daily_seasonality=False)

m.fit(df_prophet.iloc[:-6])

# Fixed future dataframe
future = m.make_future_dataframe(periods=6, freq='ME')  # 'ME' instead of 'M'
forecast_prophet = m.predict(future)
forecast_prophet_tail = forecast_prophet['yhat'].tail(6).values

print("✅ Prophet Forecast done!")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

plt.figure(figsize=(14, 7))

plt.plot(train.index, train['Sales'], label='Training Data', color='blue')
plt.plot(test.index, test['Sales'], label='Actual Test Data', color='green', marker='o')
plt.plot(test.index, forecast_arima, label='ARIMA Forecast', color='red', linestyle='--')
plt.plot(test.index, forecast_prophet_tail, label='Prophet Forecast', color='orange', linestyle='--')

plt.title('Predictive Analytics - Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Model Evaluation
print("=== Model Performance ===")
print("ARIMA MAE: ", round(mean_absolute_error(test['Sales'], forecast_arima), 2))
print("Prophet MAE:", round(mean_absolute_error(test['Sales'], forecast_prophet_tail), 2))
print("\n✅ Project visualization complete!")
```
